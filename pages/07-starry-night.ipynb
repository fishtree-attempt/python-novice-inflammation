```python
from PIL import Image
```

First we need to read the image in. 

```python
im = Image.open('starry_night.jpg')
```

If we want to display the image in our notebook, we can use functions built into IPython itself to do so. Unfortunately, the name of the module that displays images in an IPython notebook and the one we've used to read the jpeg are the same: Image. So we'll import the one from IPython and give it another name (ipyIm).

```python
from IPython.core.display import Image as ipyIm
ipyIm(filename='starry_night.jpg') 
```

```python
im.size
```

```python
im = im.resize((125, 100), Image.ANTIALIAS)
```

```python
im.size
```

```python
from ipythonblocks import BlockGrid
grid = BlockGrid(125, 100, block_size=4, lines_on=False)
```

```python
row = 0
column = 0

for pixel in im.getdata():
    grid[row,column]=pixel
    if column == 124:
        column = 0
        row = row + 1
    else:
        column = column + 1
```

```python
grid
```

```python
grid2 = BlockGrid(125, 100, block_size=4, lines_on=False)

row = 0
column = 0

for pixel in im.getdata():
    grid2[row,column].red = pixel[0]/50*50
    grid2[row,column].green = pixel[1]/50*50
    grid2[row,column].blue = pixel[2]/50*50
    if column == 124:
        column = 0
        row = row + 1
    else:
        column = column + 1
```
